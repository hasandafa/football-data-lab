## ğŸ“¦ Setup & Data Loading

Let's load match data and analyze tactical patterns like PROPER Sport Scientists do.

Not like United who probably use a whiteboard with "attack" and "defend" written on it.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Liverpool colors
LIVERPOOL_RED = '#C8102E'
LIVERPOOL_GOLD = '#F6EB61'
LIVERPOOL_TEAL = '#00B2A9'

pd.set_option('display.max_columns', None)

print("âš½ Tactical Analysis System: ACTIVATED")
print("ğŸ”´ Klopp's Gegenpressing Model: LOADED")
print("ğŸ¤¡ United's 'Vibes FC' Strategy: BLOCKED")
```

```python
# Load data
DATA_DIR = Path('../data/raw')

matches = pd.read_csv(DATA_DIR / 'matches_2024_25.csv')
league_table = pd.read_csv(DATA_DIR / 'league_table_2024_25.csv')
clubs = pd.read_csv(DATA_DIR / 'clubs.csv')
players = pd.read_csv(DATA_DIR / 'players.csv')

# Merge home team tactical info
home_clubs = clubs[['club_id', 'preferred_formation', 'playing_style']].copy()
home_clubs.columns = ['home_club_id', 'home_formation', 'home_style']
matches = matches.merge(home_clubs, on='home_club_id', how='left')

# Merge away team tactical info
away_clubs = clubs[['club_id', 'preferred_formation', 'playing_style']].copy()
away_clubs.columns = ['away_club_id', 'away_formation', 'away_style']
matches = matches.merge(away_clubs, on='away_club_id', how='left')

# Also rename goal columns to match notebook expectations
matches.rename(columns={'home_goals': 'home_score', 'away_goals': 'away_score'}, inplace=True)

print("âœ… Data loaded successfully!")
print(f"\nâš½ Match Statistics:")
print(f"   Total Matches: {len(matches):,}")
print(f"   Total Goals: {matches['home_score'].sum() + matches['away_score'].sum():,}")
print(f"   Average Goals per Match: {(matches['home_score'].sum() + matches['away_score'].sum()) / len(matches):.2f}")
print(f"\nğŸ’¡ This is PROPER tactical data analysis")
print(f"   United: 'We just play and hope for the best'")
```

## âš½ 1. Formation Effectiveness Analysis

Let's see which formations actually WIN matches!

**Liverpool's approach:** Data-driven formation selection based on opponent
**United's approach:** "What formation did we use last week? Let's change it!" 

```python
# Formation analysis
print("="*100)
print("âš½ FORMATION EFFECTIVENESS ANALYSIS")
print("="*100)

# Merge league table to get final positions
clubs_with_position = clubs.merge(league_table[['club_id', 'position']], on='club_id')

# Formation performance
formation_stats = clubs_with_position.groupby('preferred_formation').agg({
    'position': 'mean',
    'club_id': 'count'
}).round(2)
formation_stats.columns = ['Avg League Position', 'Teams Using']
formation_stats = formation_stats.sort_values('Avg League Position')

print("\nğŸ“Š Formation Performance (Lower position = Better):")
print(formation_stats)

# Most successful formations
print("\nğŸ† Most Successful Formations:")
print("   (Based on average league position)")
for idx, (formation, row) in enumerate(formation_stats.head(3).iterrows(), 1):
    print(f"   {idx}. {formation}: Avg position {row['Avg League Position']:.1f} ({int(row['Teams Using'])} teams)")

print("\nğŸ’¡ Liverpool typically uses 4-3-3 or 4-2-3-1 (flexible, attacking)")
print("ğŸ¤¡ United uses... whatever formation panic demands that week")
```

## ğŸ¨ 2. Playing Style Analysis

**Playing styles:**
- **Attacking:** High risk, high reward (Liverpool style)
- **Possession:** Control the game (also Liverpool)
- **Defensive:** Park the bus (United under pressure)
- **Balanced:** Boring but safe
- **Counter-attacking:** Fast transitions

Let's see what works!

```python
# Playing style analysis
style_stats = clubs_with_position.groupby('playing_style').agg({
    'position': 'mean',
    'club_id': 'count'
}).round(2)
style_stats.columns = ['Avg League Position', 'Teams Using']
style_stats = style_stats.sort_values('Avg League Position')

print("="*100)
print("ğŸ¨ PLAYING STYLE EFFECTIVENESS")
print("="*100)
print("\nğŸ“Š Style Performance:")
print(style_stats)

print("\nğŸ† Most Successful Playing Styles:")
for idx, (style, row) in enumerate(style_stats.head(3).iterrows(), 1):
    print(f"   {idx}. {style}: Avg position {row['Avg League Position']:.1f}")

print("\nğŸ’¡ Liverpool's success: High-intensity attacking football")
print("   Combine possession with gegenpressing")
print("   Result: Dominate games, win trophies")
print("\nğŸ¤¡ United's confusion: No clear identity")
print("   Try everything, master nothing")
print("   Result: 8th place, tactical chaos")
```

## ğŸ  3. Home Advantage Analysis

How much does playing at home actually matter?

**Liverpool at Anfield:** Fortress, opponents fear it
**United at Old Trafford:** Used to be intimidating, now... meh

```python
matches.columns
```

```python
# Home advantage analysis
matches['home_win'] = (matches['home_score'] > matches['away_score']).astype(int)
matches['away_win'] = (matches['away_score'] > matches['home_score']).astype(int)
matches['draw'] = (matches['home_score'] == matches['away_score']).astype(int)
home_wins = matches['home_win'].mean()
away_wins = matches['away_win'].mean()
draws = matches['draw'].mean()

avg_home_goals = matches['home_score'].mean()
avg_away_goals = matches['away_score'].mean()

print("="*100)
print("ğŸ  HOME ADVANTAGE ANALYSIS")
print("="*100)
print(f"\nğŸ“Š Overall Statistics:")
print(f"   Home Win Rate: {home_wins*100:.1f}%")
print(f"   Away Win Rate: {away_wins*100:.1f}%")
print(f"   Draw Rate: {draws*100:.1f}%")
print(f"\nâš½ Goals:")
print(f"   Average Home Goals: {avg_home_goals:.2f}")
print(f"   Average Away Goals: {avg_away_goals:.2f}")
print(f"   Home Advantage: +{(avg_home_goals - avg_away_goals):.2f} goals per match")

print(f"\nğŸ’¡ Home advantage is REAL: {((home_wins - away_wins) / away_wins * 100):.0f}% more likely to win at home")
print(f"ğŸ† Liverpool maximizes this with Anfield atmosphere + crowd as 12th man")
print(f"ğŸ¤¡ United fans leave Old Trafford early when losing, nullifying advantage")
```

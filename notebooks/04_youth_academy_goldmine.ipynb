## üìà 2. Growth Trajectory Analysis

Let's visualize how these youth players will develop over time.

Without them (United style), they'll stagnate and leave for free.

```python
# Growth trajectory visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Current vs Potential Rating
axes[0, 0].scatter(youth['overall_rating'], youth['potential'], 
                  alpha=0.6, s=80, c=youth['age'], cmap='RdYlGn_r',
                  edgecolors='black', linewidth=0.5)

# Add diagonal line (no growth)
min_val = min(youth['overall_rating'].min(), youth['potential'].min())
max_val = max(youth['overall_rating'].max(), youth['potential'].max())
axes[0, 0].plot([min_val, max_val], [min_val, max_val], 
               'k--', alpha=0.5, linewidth=2, label='No Growth Line')

# Highlight elite potential zone
axes[0, 0].axhspan(80, 100, alpha=0.1, color='green', label='Elite Potential (80+)')
axes[0, 0].axhspan(85, 100, alpha=0.15, color='gold', label='World-Class (85+)')

axes[0, 0].set_xlabel('Current Overall Rating', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Potential Rating', fontsize=12, fontweight='bold')
axes[0, 0].set_title('Youth Development Potential\n(Further from line = More growth potential)', 
                     fontsize=13, fontweight='bold', color=LIVERPOOL_RED)
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)
plt.colorbar(axes[0, 0].collections[0], ax=axes[0, 0], label='Age')

# Plot 2: Potential Growth by Position
position_growth = youth.groupby('position_group')['potential_growth'].mean().sort_values(ascending=False)
colors_pos = [LIVERPOOL_RED, LIVERPOOL_TEAL, LIVERPOOL_GOLD, 'coral']
bars = axes[0, 1].bar(range(len(position_growth)), position_growth.values, 
                     color=colors_pos[:len(position_growth)], alpha=0.8, edgecolor='black', linewidth=1.5)

axes[0, 1].set_xticks(range(len(position_growth)))
axes[0, 1].set_xticklabels(position_growth.index, fontsize=11)
axes[0, 1].set_ylabel('Average Potential Growth', fontsize=12, fontweight='bold')
axes[0, 1].set_title('Growth Potential by Position\n(Higher = More room to develop)', 
                     fontsize=13, fontweight='bold', color=LIVERPOOL_RED)
axes[0, 1].grid(axis='y', alpha=0.3)

# Add value labels
for i, bar in enumerate(bars):
    height = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width()/2., height,
                   f'+{height:.1f}',
                   ha='center', va='bottom', fontweight='bold')

# Plot 3: Age Distribution
age_counts = youth['age'].value_counts().sort_index()
axes[1, 0].bar(age_counts.index, age_counts.values, color=LIVERPOOL_RED, alpha=0.7, edgecolor='black')
axes[1, 0].set_xlabel('Age', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Number of Players', fontsize=12, fontweight='bold')
axes[1, 0].set_title('Youth Academy Age Distribution\n(Younger = More development time)', 
                     fontsize=13, fontweight='bold', color=LIVERPOOL_RED)
axes[1, 0].grid(axis='y', alpha=0.3)

# Plot 4: Potential Distribution
axes[1, 1].hist(youth['potential'], bins=20, color=LIVERPOOL_RED, alpha=0.7, edgecolor='black')
axes[1, 1].axvline(80, color='green', linestyle='--', linewidth=2, label='Elite Threshold (80)')
axes[1, 1].axvline(85, color='gold', linestyle='--', linewidth=2, label='World-Class (85)')
axes[1, 1].axvline(youth['potential'].mean(), color=LIVERPOOL_TEAL, linestyle='-', 
                  linewidth=3, label=f'Average: {youth["potential"].mean():.1f}')
axes[1, 1].set_xlabel('Potential Rating', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Number of Players', fontsize=12, fontweight='bold')
axes[1, 1].set_title('Potential Rating Distribution\n(Right side = Future stars)', 
                     fontsize=13, fontweight='bold', color=LIVERPOOL_RED)
axes[1, 1].legend()
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../data/processed/youth_growth_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("üìä Growth trajectory analysis complete!")
print("üí° Players above the diagonal line have HIGH growth potential")
print("üéØ Focus development resources on players with potential 80+")
```

## ‚ö° 3. Promotion Readiness Assessment

When should you promote youth to the first team?

**Liverpool's approach:** Age 18-19 if rating 70+, patient development
**United's approach:** Rush them in at 17, destroy confidence, sell them, cry later

Let's identify who's ready for promotion!

```python
# Promotion readiness criteria
print("="*100)
print("‚ö° PROMOTION READINESS ASSESSMENT")
print("="*100)

# Define readiness categories
youth['promotion_status'] = 'Not Ready'

# Ready for first team (rating 70+, age 18+)
youth.loc[(youth['overall_rating'] >= 70) & (youth['age'] >= 18), 'promotion_status'] = 'READY NOW'

# Almost ready (rating 65-70, age 18+)
youth.loc[(youth['overall_rating'] >= 65) & (youth['overall_rating'] < 70) & (youth['age'] >= 18), 
          'promotion_status'] = 'Almost Ready (6-12 months)'

# High potential, needs time (potential 80+, age 16-17)
youth.loc[(youth['potential'] >= 80) & (youth['age'] <= 17), 
          'promotion_status'] = 'Future Star (2-3 years)'

# Good loan candidates (rating 65+, age 18+, not quite first team)
youth.loc[(youth['overall_rating'] >= 65) & (youth['overall_rating'] < 70) & 
          (youth['age'] >= 18) & (youth['potential'] >= 75),
          'promotion_status'] = 'Loan Candidate'

status_counts = youth['promotion_status'].value_counts()
print("\nüìä Promotion Readiness Breakdown:")
for status, count in status_counts.items():
    pct = count / len(youth) * 100
    print(f"   {status}: {count} players ({pct:.1f}%)")

print("\nüí° DEVELOPMENT STRATEGY:")
print("   ‚úÖ READY NOW: Integrate into first team squad immediately")
print("   ‚úÖ Almost Ready: Train with first team, play in cup games")
print("   ‚úÖ Loan Candidate: Send on loan to get regular playing time")
print("   ‚úÖ Future Star: Keep in academy, monitor closely")
```

## üí∞ 5. Academy Return on Investment (ROI)

Youth academies are **MASSIVE profit centers** if run properly.

**Liverpool's academy ROI:**
- Trent Alexander-Arnold: ¬£0 cost ‚Üí ¬£80M+ value = INFINITE ROI
- Sterling (sold): ¬£0 cost ‚Üí ¬£49M sale = INFINITE ROI  
- **Total academy value:** Hundreds of millions

**United's academy disaster:**
- Paul Pogba: Let go for ¬£0 ‚Üí Bought back ¬£89M = -¬£89M ROI ü§°
- Lost countless talents due to poor development

Let's calculate YOUR academy's value!

```python
# Calculate academy ROI
print("="*100)
print("üí∞ YOUTH ACADEMY ROI ANALYSIS")
print("="*100)

# Estimate market value for youth based on potential
# Youth aren't officially valued, but we can estimate based on potential
def estimate_youth_value(potential, age):
    '''Estimate youth player market value based on potential'''
    if potential >= 90:
        base = 15_000_000  # ¬£15M for wonderkids
    elif potential >= 85:
        base = 8_000_000   # ¬£8M for world-class potential
    elif potential >= 80:
        base = 4_000_000   # ¬£4M for elite potential
    elif potential >= 75:
        base = 2_000_000   # ¬£2M for strong potential
    else:
        base = 500_000     # ¬£500k for average

    # Younger = more valuable (more development time)
    age_multiplier = 1.2 if age <= 16 else 1.0
    
    return base * age_multiplier

youth['estimated_value'] = youth.apply(lambda x: estimate_youth_value(x['potential'], x['age']), axis=1)
youth['estimated_value_millions'] = youth['estimated_value'] / 1_000_000

total_academy_value = youth['estimated_value'].sum()
avg_value_per_player = youth['estimated_value'].mean()

print(f"\nüíé ACADEMY ASSET VALUE:")
print(f"   Total Estimated Value: ¬£{total_academy_value/1_000_000:.1f}M")
print(f"   Average per Player: ¬£{avg_value_per_player/1_000_000:.2f}M")
print(f"   Number of Players: {len(youth)}")

print(f"\nüí∞ DEVELOPMENT COST vs VALUE:")
print(f"   Estimated Total Development Cost: ¬£{len(youth) * 50_000:,} (¬£50k per player)")
print(f"   Total Estimated Value: ¬£{total_academy_value:,.0f}")
print(f"   ROI: {(total_academy_value / (len(youth) * 50_000) - 1) * 100:.0f}%")

print(f"\nüåü ELITE PROSPECTS VALUE:")
elite = youth[youth['potential'] >= 80]
if len(elite) > 0:
    elite_value = elite['estimated_value'].sum()
    print(f"   Elite Prospects (80+ potential): {len(elite)} players")
    print(f"   Combined Value: ¬£{elite_value/1_000_000:.1f}M")
    print(f"   Average Value: ¬£{elite_value/len(elite)/1_000_000:.1f}M each")

print(f"\nüèÜ THE LIVERPOOL MODEL:")
print(f"   Academy Investment: ¬£5M per year")
print(f"   Produce 2-3 first team players per year")
print(f"   Value created: ¬£50M+ per year")
print(f"   NET PROFIT: ¬£45M+ per year (WITHOUT selling!)")

print(f"\nü§° THE UNITED MODEL:")
print(f"   Academy Investment: ¬£10M per year (waste)")
print(f"   Produce 0-1 first team players per year")
print(f"   Let talents go for free")
print(f"   Buy worse players for ¬£80M+")
print(f"   NET LOSS: ¬£70M+ per year (buying back your own players!)")
```

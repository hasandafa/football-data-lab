## üíé FINDING GEMS: THE UNDERVALUED MARKET

This is where Liverpool made their fortune. Finding players who are:
- ‚úÖ High quality
- ‚úÖ Relatively cheap  
- ‚úÖ Have room to improve

You know, the exact OPPOSITE of signing Harry Maguire for ¬£80M.

Let's find our Robertson (¬£8M), our Salah (¬£36M), our Firmino (¬£29M)...

```python
# Calculate value score (rating per million)
players['value_score'] = players['overall_rating'] / (players['market_value'] / 1e6)
players['potential_growth'] = players['potential'] - players['overall_rating']

# Find undervalued gems (high rating, low cost, high potential)
gems = players[
    (players['overall_rating'] >= 75) &  # Good quality
    (players['market_value'] <= 15e6) &  # Affordable
    (players['potential_growth'] >= 5) &  # Room to improve
    (players['age'] <= 26)  # Still young enough
].copy()

print("üíé HIDDEN GEMS: THE LIVERPOOL SCOUTING METHOD\n" + "="*80)
print(f"Found {len(gems)} potential bargains\n")
print("Criteria:")
print("  ‚úÖ Rating ‚â• 75 (Quality)")
print("  ‚úÖ Value ‚â§ ¬£15M (Affordable)")
print("  ‚úÖ Potential growth ‚â• 5 (Room to improve)")
print("  ‚úÖ Age ‚â§ 26 (Prime years ahead)\n")

if len(gems) > 0:
    # Sort by value score
    gems_display = gems.nlargest(15, 'value_score')[[
        'full_name', 'age', 'nationality', 'primary_position',
        'overall_rating', 'potential', 'market_value', 'value_score'
    ]].copy()
    
    gems_display['market_value'] = gems_display['market_value'].apply(lambda x: f"¬£{x/1e6:.1f}M")
    gems_display['value_score'] = gems_display['value_score'].round(2)
    
    print("üåü TOP 15 BARGAIN PLAYERS:\n")
    print(gems_display.to_string(index=False))
    
    print("\n" + "="*80)
    print("üí∞ WHAT LIVERPOOL WOULD DO:")
    print("   1. Sign these players for combined ¬£150M")
    print("   2. Develop them properly with sport scientists")
    print("   3. Win multiple trophies")
    print("   4. Sell the ones who don't fit for profit")
    print("\nüíÄ WHAT UNITED WOULD DO:")
    print("   1. Ignore all of them")
    print("   2. Spend ¬£150M on ONE player who had a good World Cup")
    print("   3. Watch him flop")
    print("   4. Finish 8th")
    print("   5. Blame the manager")
    print("   6. Repeat")
else:
    print("No gems found with these criteria (try adjusting parameters)")
```

```python
# Create market_value_millions column (convert from raw value)
players['market_value_millions'] = players['market_value'] / 1_000_000

print(f"‚úÖ Market value converted to millions")
print(f"üìä Range: ¬£{players['market_value_millions'].min():.2f}M - ¬£{players['market_value_millions'].max():.2f}M")

# Visualize value vs performance
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Plot 1: Market Value vs Overall Rating
scatter = axes[0].scatter(players['market_value_millions'], players['overall_rating'], 
                         alpha=0.6, s=60, c=players['age'], cmap='RdYlGn_r',
                         edgecolors='black', linewidth=0.5)
cbar = plt.colorbar(scatter, ax=axes[0], label='Age (Years)')

# REALISTIC LIVERPOOL ZONE (based on actual Liverpool transfers)
# Liverpool targets: Age 24-28, Rating 82-88, Price ¬£30-80M
liverpool_zone_x = [30, 80, 80, 30, 30]  # ¬£30-80M
liverpool_zone_y = [82, 82, 88, 88, 82]  # Rating 82-88
axes[0].fill(liverpool_zone_x, liverpool_zone_y, color='green', alpha=0.15, zorder=1)
axes[0].text(55, 85, 'LIVERPOOL\nZONE', fontsize=11, ha='center', fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='green', alpha=0.3, edgecolor='darkgreen', linewidth=2))

# REALISTIC UNITED DISASTER ZONE (Maguire ¬£80M, Antony ¬£85M type deals)
# Overpaid: ¬£60-100M for rating 75-82
united_zone_x = [60, 100, 100, 60, 60]  # ¬£60-100M
united_zone_y = [75, 75, 82, 82, 75]    # Rating 75-82
axes[0].fill(united_zone_x, united_zone_y, color='red', alpha=0.12, zorder=1)
axes[0].text(80, 78.5, 'UNITED\nZONE\n(Overpaid)', fontsize=11, ha='center', fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='red', alpha=0.3, edgecolor='darkred', linewidth=2))

# Add reference lines
axes[0].axhline(85, color='gold', linestyle='--', alpha=0.5, linewidth=2, label='Elite Rating (85+)', zorder=2)
axes[0].axvline(100, color='purple', linestyle='--', alpha=0.5, linewidth=2, label='Premium Price (¬£100M+)', zorder=2)

axes[0].set_xlabel('Market Value (¬£M)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Overall Rating', fontsize=12, fontweight='bold')
axes[0].set_title('Market Value vs Overall Rating\n(Color: Green=Young | Red=Old)', 
                 fontsize=13, fontweight='bold', color=LIVERPOOL_RED)
axes[0].legend(loc='lower right', fontsize=9)
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim(-5, 220)
axes[0].set_ylim(45, 100)

# Plot 2: Best Value by Position (FIXED - Prime Age Only)
prime_players = players[(players['age'] >= 24) & (players['age'] <= 28)].copy()

if len(prime_players) > 0:
    # Get best value player per position
    top_value_by_pos = prime_players.loc[
        prime_players.groupby('position_group')['value_score'].idxmax()
    ]
    
    positions = top_value_by_pos['position_group'].values
    values = top_value_by_pos['value_score'].values
    ratings = top_value_by_pos['overall_rating'].values
    market_vals = top_value_by_pos['market_value_millions'].values
    
    bars = axes[1].barh(positions, values, color=LIVERPOOL_RED, 
                       edgecolor='black', linewidth=1.5, alpha=0.8)
    
    # Add detailed labels
    for i, (pos, val, rating, mv) in enumerate(zip(positions, values, ratings, market_vals)):
        label_text = f'{rating:.0f} rating | ¬£{mv:.0f}M'
        axes[1].text(val + 1, i, label_text, 
                    va='center', fontsize=9, fontweight='bold')
    
    axes[1].set_xlabel('Value Score (Rating per ¬£M)', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Position', fontsize=12, fontweight='bold')
    axes[1].set_title('Best Value Players by Position\n(Prime Age 24-28 Only)', 
                     fontsize=13, fontweight='bold', color=LIVERPOOL_RED)
    axes[1].grid(True, alpha=0.3, axis='x')
    axes[1].set_xlim(0, max(values) * 1.25)

plt.tight_layout()
plt.savefig('../data/processed/transfer_value_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nüéØ TRANSFER MARKET ZONES EXPLAINED:")
print("="*80)
print("\nüü¢ LIVERPOOL ZONE (Smart Money):")
print("   üìä Rating: 82-88 (Elite but not absurd)")
print("   üí∞ Price: ¬£30-80M (Fair value)")
print("   üë§ Age: 24-28 (Peak years)")
print("   ‚úÖ Examples: Salah ¬£37M (85), Robertson ¬£8M (84), Man√© ¬£34M (86)")
print(f"   üìà Players in this zone: {len(players[(players['overall_rating'] >= 82) & (players['overall_rating'] <= 88) & (players['market_value_millions'] >= 30) & (players['market_value_millions'] <= 80) & (players['age'] >= 24) & (players['age'] <= 28)])}")

print("\nüî¥ UNITED ZONE (Expensive Disasters):")
print("   üìä Rating: 75-82 (Good but not elite)")
print("   üí∞ Price: ¬£60-100M (MASSIVELY overpaid)")
print("   üë§ Age: Often 28+ (Declining or peak passed)")
print("   ‚ùå Examples: Maguire ¬£80M (78), Antony ¬£85M (80), Sancho ¬£73M (82)")
print(f"   üìâ Players in this zone: {len(players[(players['overall_rating'] >= 75) & (players['overall_rating'] <= 82) & (players['market_value_millions'] >= 60) & (players['market_value_millions'] <= 100)])}")

print("\nüí° THE LESSON:")
print("   Liverpool: Elite quality (82-88) at fair prices (¬£30-80M) = Smart!")
print("   United: Good quality (75-82) at stupid prices (¬£60-100M) = Disaster!")
print("\nüèÜ YNWA! Pay your Sport Scientists!")
```
